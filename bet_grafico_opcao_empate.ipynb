```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, ToggleButtons, fixed

# Função para calcular as apostas ideais para Time A e Time B sem empate
def calcular_apostas_sem_empate(odds_a, odds_b, max_investimento=100):
    investimento_a = max_investimento / (1 + (odds_b / odds_a))
    investimento_b = max_investimento - investimento_a
    return investimento_a, investimento_b

# Função para calcular as apostas ideais para Time A, Time B e Empate
def calcular_apostas_com_empate(odds_a, odds_b, odds_empate, max_investimento=100):
    investimento_a = max_investimento / (1 + (odds_b / odds_a) + (odds_empate / odds_a))
    investimento_b = (odds_b / odds_a) * investimento_a
    investimento_empate = max_investimento - (investimento_a + investimento_b)

    if investimento_empate < 0:
        investimento_empate = 0
        investimento_a = max_investimento / (1 + (odds_b / odds_a))
        investimento_b = max_investimento - investimento_a

    return investimento_a, investimento_b, investimento_empate

# Função para calcular o lucro/perda em cada cenário sem empate
def calcular_lucro_sem_empate(aposta_a, aposta_b, odds_a, odds_b):
    retorno_a = aposta_a * odds_a
    retorno_b = aposta_b * odds_b
    lucro_a = retorno_a - (aposta_a + aposta_b)
    lucro_b = retorno_b - (aposta_a + aposta_b)
    return lucro_a, lucro_b

# Função para calcular o lucro/perda em cada cenário com empate
def calcular_lucro_com_empate(aposta_a, aposta_b, aposta_empate, odds_a, odds_b, odds_empate):
    retorno_a = aposta_a * odds_a
    retorno_b = aposta_b * odds_b
    retorno_empate = aposta_empate * odds_empate
    lucro_a = retorno_a - (aposta_a + aposta_b + aposta_empate)
    lucro_b = retorno_b - (aposta_a + aposta_b + aposta_empate)
    lucro_empate = retorno_empate - (aposta_a + aposta_b + aposta_empate)
    return lucro_a, lucro_b, lucro_empate

# Função principal para simular os cenários
def simular_cenario(tipo_de_aposta, odds_a=2.16, odds_b=3.55, odds_empate=2.56):
    if tipo_de_aposta == 'Com Empate':
        aposta_a, aposta_b, aposta_empate = calcular_apostas_com_empate(odds_a, odds_b, odds_empate)
        lucro_a, lucro_b, lucro_empate = calcular_lucro_com_empate(aposta_a, aposta_b, aposta_empate, odds_a, odds_b, odds_empate)

        retorno_liquido_a = aposta_a * odds_a
        retorno_liquido_b = aposta_b * odds_b
        retorno_liquido_empate = aposta_empate * odds_empate

        # Exibindo os resultados em três colunas
        print(f"{'Cenário':<20}{'Aposta no Time (R$)':<25}{'Retorno Líquido (R$)':<25}{'Lucro se o Time (R$)'}")
        print("-" * 80)
        print(f"{'Vitória Time A':<20}R${aposta_a:<23.2f}R${retorno_liquido_a:<23.2f}R${lucro_a:.2f}")
        print(f"{'Vitória Time B':<20}R${aposta_b:<23.2f}R${retorno_liquido_b:<23.2f}R${lucro_b:.2f}")
        print(f"{'Empate':<20}R${aposta_empate:<23.2f}R${retorno_liquido_empate:<23.2f}R${lucro_empate:.2f}")

        # Gráfico dos lucros/perdas
        resultados = ['Vitória Time A', 'Empate', 'Vitória Time B']
        lucros = [lucro_a, lucro_empate, lucro_b]

    else:
        aposta_a, aposta_b = calcular_apostas_sem_empate(odds_a, odds_b)
        lucro_a, lucro_b = calcular_lucro_sem_empate(aposta_a, aposta_b, odds_a, odds_b)

        retorno_liquido_a = aposta_a * odds_a
        retorno_liquido_b = aposta_b * odds_b

        # Exibindo os resultados em três colunas
        print(f"{'Cenário':<20}{'Aposta no Time (R$)':<25}{'Retorno Líquido (R$)':<25}{'Lucro se o Time (R$)'}")
        print("-" * 70)
        print(f"{'Vitória Time A':<20}R${aposta_a:<23.2f}R${retorno_liquido_a:<23.2f}R${lucro_a:.2f}")
        print(f"{'Vitória Time B':<20}R${aposta_b:<23.2f}R${retorno_liquido_b:<23.2f}R${lucro_b:.2f}")

        # Gráfico dos lucros/perdas
        resultados = ['Vitória Time A', 'Vitória Time B']
        lucros = [lucro_a, lucro_b]

    plt.figure(figsize=(5, 3))
    plt.plot(resultados, lucros, marker='o', linestyle='-', color='b')
    plt.title(f'Simulação de Estratégia de Hedge - {tipo_de_aposta}')
    plt.xlabel('Cenário')
    plt.ylabel('Lucro/Perda (R$)')
    plt.grid(True)
    plt.axhline(0, color='r', linestyle='--')
    plt.show()

# Caixa de seleção para escolher o tipo de aposta
tipo_de_aposta_widget = ToggleButtons(
    options=['Com Empate', 'Sem Empate'],
    description='Tipo de Aposta:',
)

# Interatividade com sliders e a caixa de seleção para ajustar as odds
def criar_interface(tipo_de_aposta):
    if tipo_de_aposta == 'Com Empate':
        interact(simular_cenario, tipo_de_aposta=fixed(tipo_de_aposta),
                 odds_a=FloatSlider(value=2.16, min=1.1, max=10.0, step=0.01, description='Odds Time A'),
                 odds_b=FloatSlider(value=3.55, min=1.1, max=10.0, step=0.01, description='Odds Time B'),
                 odds_empate=FloatSlider(value=2.56, min=1.1, max=10.0, step=0.01, description='Odds Empate'))
    else:
        interact(simular_cenario, tipo_de_aposta=fixed(tipo_de_aposta),
                 odds_a=FloatSlider(value=1.90, min=1.1, max=10.0, step=0.01, description='Odds Time A'),
                 odds_b=FloatSlider(value=2.10, min=1.1, max=10.0, step=0.01, description='Odds Time B'))

interact(criar_interface, tipo_de_aposta=tipo_de_aposta_widget)

```
